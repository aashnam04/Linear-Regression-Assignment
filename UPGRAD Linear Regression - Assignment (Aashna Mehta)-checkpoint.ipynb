## Step 5 : Model Building

```python
lm = LinearRegression()
lm.fit(x_train, y_train)
```

```python
#Recursive Feature Elimination
rfe = RFE(lm,n_features_to_select=15)
rfe = rfe.fit(x_train, y_train)
```

```python
#list of varibales selected in the top 15 list
list(zip(x_train.columns,rfe.support_,rfe.ranking_))
```

```python
#selecting the selected variable via RFE in col list
cols = x_train.columns[rfe.support_]
print(cols)
```

```python
#checking which columns has been rejected
x_train.columns[~rfe.support_]
```

###### Building regression models and then dropping the variables after looking at the p-values and VIF 

```python
x_train_rfe = x_train[cols]
```

```python
x_train_rfe = sm.add_constant(x_train_rfe)
```

```python
lm = sm.OLS(y_train,x_train_rfe).fit()
```

```python
print(lm.summary())
```

```python
def calculateVIF(df):
    vif = pd.DataFrame()
    vif['Features'] = df.columns
    vif['VIF'] = [variance_inflation_factor(df.values, i) for i in range(df.shape[1])]
    vif['VIF'] = round(vif['VIF'], 2)
    vif = vif.sort_values(by = "VIF", ascending = False)
    return vif 
```

```python
x_train_rfe = x_train[cols]
```

```python
calculateVIF(x_train_rfe)
```

```python
lr_model = lr.fit()
```

```python
lr_model.summary()
```

## Step 6 : Residual Analysis

```python
y_train_pred = lr_model.predict(x_train_sm_3)
res = (y_train - y_train_pred)
```

```python
#plot the histogram of the error terms
plt.figure(figsize = (12,6))
sns.set_style("darkgrid")
sns.distplot(res)
plt.title('Distribution of Error Terms')
plt.show()
```

We can see Error Terms are normally Distributed with mean Zero. Hence Model is Actaully obeying the assumption of Normality.

```python
#looking for patterns in the residuals
plt.figure(figsize = (12,6))
sns.set_style("darkgrid")
plt.scatter(y_train_pred,res)
plt.title('Residual Vs. Predicted Values')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.show()
```

We can see there is no specific Pattern observed in the Error Terms with respect to Prediction, hence we can say Error terms areindependentof each other.

```python
#looking for constant variance
plt.figure(figsize = (12,6))
sns.set_style("darkgrid")
sns.regplot(x=y_train, y=y_train_pred)
plt.title('Predicted Points Vs. Actual Points')
plt.xlabel('Actual Points')
plt.ylabel('Predicted Points')
plt.show()
```

We can see Error Terms have approximatly a Constant Variance, Hence it follows the Assumption of Homoscedasticity
